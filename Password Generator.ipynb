```python
#@title Simple Random Password Generator

lower_case = 'abcdefghijklmnopqrstuvwxyz'  #@param {type: "string"}
upper_case = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'  #@param {type: "string"}
symbol = '!@#$%%^&*()_=+-'  #@param {type: "string"}
number = '0123456789'  #@param {type: "string"}
length = 23 #@param {type:"slider", min:8, max:65, step:1}


import random 

data = lower_case + upper_case + symbol + number

# in case of value error 
if (length > len(data)):
  print("ValueError for length : ",length,' so it was changed to : ', len(data))
  length = len(data)


password = ''.join(random.sample(data,length))

print(password)
```
